```python
import pandas as pd
```

```python
songs = pd.read_csv('charts.csv')
songs
```

```python
not_hot_songs = songs.sample(n=3000, random_state=0)
```

```python
not_hot_songs = not_hot_songs.reset_index()
```

```python
not_hot_songs.drop(['index'], axis=1)
```

```python
not_hot_songs_sliced = not_hot_songs.loc[:,['title', 'artist', 'url']]
```

```python
not_hot_songs_sliced
```

```python
top_hot_song = pd.read_csv('../../Morning/lab-web-scraping-single-page/top_hot_songs.csv')
top_hot_song
```

```python
def title_matcher(column1, column2):
    for value1 in column1:
        if value1 in column2:
            return 'There is a match: ', value1
        else:
            return 'No match'
```

```python
title_matcher(not_hot_songs_sliced['title'], top_hot_song['title'])
```
