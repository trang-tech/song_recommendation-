```python
def get_song_id(song_name, artist_name):
    try: 
        results=sp.search(q='tracks:' +song_name+"artist:"+artist_name,limit = 1)
        return results['tracks']['items'][0]['id']
    except: 
        return "None"
```

```python
def add_audio_features(title_chunked, artist_chunked):
    list_of_ids = []
    all_my_dict = {}
    for tit, art in zip(title_chunked, artist_chunked):    
        for t, a in zip(tit, art):
            new_id = get_song_id(t, a)
            #list_of_ids.append(new_id)
            #print(new_id)
            if new_id != "None":
                try:
                    my_dict = sp.audio_features(new_id)[0]
                    all_my_dict[new_id] = my_dict
                    all_my_dict[new_id]['song_name'] = t  # add song name to the dictionary
                    all_my_dict[new_id]['artist_name'] = a  # add artist name to the dictionary
                except:
                    print ("Couldn't get the audio features")
            
            else:
                print("Invalid URI ({}, {})".format(a, t))
            
        print('I will sleep for 10 seconds')
        sleep(10)
        
    df = pd.DataFrame(all_my_dict).T
    return df
```
